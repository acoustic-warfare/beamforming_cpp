```python
import numpy as np
import math
import soundfile as sf
from scipy import signal
import config

# GLOBAL VARIABLES
c = 343
```

```python


# Initialization
# f_sampling, t_start, t_end, away_distance are user defined variables
f_sampling = config.f_sampling                      # sampling frequency in Hz
```

```python
# Filtering values
f_bands_N = config.f_bands_N         # number of frequency bands
#f_bands_N = 5
bandwidth = config.bandwidth         # bandwidth of incoming audio signal
frequency_bands = np.linspace(bandwidth[0],bandwidth[1],f_bands_N) # vector holding center frequencies of all frequency bands
```

```python
def calculate_filter_coefficients(f_sampling, frequency_bands):
    scale_factor = config.scale_factor
    f_bands_N = config.f_bands_N
    filter_order = config.filter_order
    f_coefficients = np.zeros((f_bands_N, filter_order))
    for freq_ind in range(config.f_bands_N):
        nu_0 = 2*frequency_bands[freq_ind]/f_sampling
        cut_off = [nu_0 - nu_0/scale_factor, nu_0 + nu_0/scale_factor]
        b = signal.firwin(filter_order, cut_off, window="hamming", pass_zero=False) # filter coefficients
        f_coefficients[freq_ind,:] = b
    return f_coefficients
```

```python
filter_coefficients = calculate_filter_coefficients(f_sampling, frequency_bands)

#np.savetxt('filter_coeff_pyython.txt',filter_coefficients,fmt='%.18f' , delimiter=',') 

f = open("filter_coefficients.h", "w")
f.write("#ifndef FILTER_COEFFICIENTS_H \n")
f.write("#define FILTER_COEFFICIENTS_H \n")
f.write("\n")
f.write("namespace filter_coefficients \n")
f.write("{ \n")
f.write("\t// Number of frequency bands generated by generate_filter_coefficients_py.ipynb \n")
f.write("\tconstexpr int f_bands_N { " + str(config.f_bands_N) + " };\n\n")

f.write("\t// Filter order generated by generate_filter_coefficients_py.ipynb \n")
f.write("\tconstexpr int filter_order { " + str(config.filter_order) + " };\n\n")

f.write("\t// Center frequencies generated by generate_filter_coefficients_py.ipynb \n")
f.write("\tconstexpr double center_frequencies[" + str(config.f_bands_N) + "] { ")

for freq_ind in range(len(frequency_bands)):
    f.write(str(frequency_bands[freq_ind]))
    if freq_ind != config.f_bands_N-1:
            f.write(", ")


f.write(" };\n\n" )
f.write("\t// FIlter coefficients generated by generate_filter_coefficients_py.ipynb \n")
f.write("\tconstexpr double filt_coeffs[" + str(config.f_bands_N) + "][" + str(config.filter_order) + "]\n \t" )
f.write("{")

for freq_ind in range(config.f_bands_N):
    f.write("\n\t\t{")
    for b_coeff in range(config.filter_order):
        f.write(str(filter_coefficients[freq_ind,b_coeff]))
        if b_coeff != config.filter_order-1:
            f.write(", ")
    f.write("\n\t\t}")
    if freq_ind != config.f_bands_N-1:
        f.write(", ")

f.write("\n\t};")
f.write("\n}")
f.write("\n#endif")
f.close()

```
